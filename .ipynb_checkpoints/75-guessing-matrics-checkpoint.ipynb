# Alpha guesswork
- Question: How can we model how hard it would be for an attacker to guess this dataset?
- Expectation Calculation: Guess each password in probability order
how many guesses would it take to guess any randomly selected password?
![image](guessing-matrics.png)

# Testing the code with in class data

```python
#  prob is a probability distribution (of how humans choose passwords)
prob = []
def probability(count, total):
    for i in count:
        p = round(i/total, 4)
        prob.append(p)
    return prob
```

```python
# Percent guessed after ùõΩ guesses
def beta_success_rate(df):
    b_success_rate = []
    for b in range(len(df)):
        rate = df.loc[0:b,'prob.'].sum()
        b_success_rate.append(round(rate,3))
    df['beta'] = np.arange(1, len(df)+1)
    df['beta_success_rate'] = b_success_rate
    return df
```

```python
# Number of guesses per account to crack a desired fraction
def random_guess(df):
    df['random_guess'] = df['beta'] * df['prob.']
    return df
```

```python
# Proportion not guessed after ùúá guesses 
# Plus
# Expected number of guesses to guess a random password among the first ùúá passwords
def guesswork(alpha,df):
    for i in range(len(df)):
        if df['beta_success_rate'][i] <= alpha:
            alpha_wf = df['beta'][i+1]
    sum_random_guess = df.loc[0:alpha_wf,'random_guess'].sum()
    g = (1 - df['beta_success_rate'][alpha_wf]) * alpha_wf + sum_random_guess
    return g
```

```python
# test case
data = {'count':[50,42, 30, 29, 26, 15, 14, 14, 13, 12, 9, 6,3,1,1,1],
        'password': ['123456','123456789' ,'password','iloveyou', 'princess','abc123',
                    'Nicole','Daniel','babygirl','monkey','lovely','Jessica',
                     '654321','Michael','Ashley','qwerty']}
# Create DataFrame
example = pd.DataFrame(data)
```

```python
example['prob.'] = probability(example['count'], example['count'].sum())
```

```python
example = beta_success_rate(example)
```

```python
example = random_guess(example)
```

```python
example
```

```python
guesswork(alpha=0.5,df=example)
```

# hotmail

```python
#  prob is a probability distribution (of how humans choose passwords)
prob = []
total = hotmail['count'].sum()
for i in hotmail['count']:
    p = i/total
    prob.append(p)
hotmail['prob.'] = prob
```

```python
# Percent guessed after ùõΩ guesses

b_success_rate = []
for b in range(len(hotmail)):
    rate = hotmail.loc[0:b,'prob.'].sum()
    b_success_rate.append(rate)
hotmail['beta'] = np.arange(1, len(hotmail)+1)
hotmail['beta_success_rate'] = b_success_rate
```

```python
# Number of guesses per account to crack a desired fraction
hotmail['random_guess'] = hotmail['beta'] * hotmail['prob.']
```

```python

# Proportion not guessed after ùúá guesses 
# Plus
# Expected number of guesses to guess a random password among the first ùúá passwords
alpha = [0.05, 0.10, 0.15, 0.20, 0.25, 0.50, 0.75, 0.95]
result_hotmail = []
for a in alpha:
    for i in range(len(hotmail)):
        if hotmail['beta_success_rate'][i] <= a:
            alpha_wf = hotmail['beta'][i+1]
    sum_random_guess = hotmail.loc[0:alpha_wf,'random_guess'].sum()
    g = (1 - hotmail['beta_success_rate'][alpha_wf]) * alpha_wf + sum_random_guess
    result_hotmail.append(g)
```

```python
#  prob is a probability distribution (of how humans choose passwords)
prob = []
total = rockyou_dropnan['count'].sum()
for i in rockyou_dropnan['count']:
    p = i/total
    prob.append(p)
rockyou_dropnan['prob.'] = prob
```

```python
rockyou_dropnan
```

```python
from itertools import accumulate
# Percent guessed after ùõΩ guesses
b_success_rate = list(accumulate(rockyou_dropnan['prob.']))

rockyou_dropnan['beta'] = np.arange(1, len(rockyou_dropnan)+1)
rockyou_dropnan['beta_success_rate'] = b_success_rate
```

```python

# Number of guesses per account to crack a desired fraction
rockyou_dropnan['random_guess'] = rockyou_dropnan['beta'] * rockyou_dropnan['prob.']
```

```python
rockyou_dropnan.head()
```

```python
df = rockyou_dropnan[rockyou_dropnan.loc[:,'random_guess'] <=  0.05 + 0.001 ]
df
```

```python

# Proportion not guessed after ùúá guesses 
# Plus
# Expected number of guesses to guess a random password among the first ùúá passwords
alpha = [0.05, 0.10, 0.15, 0.20, 0.25, 0.50, 0.75, 0.95]
result_rockyou = []
alpha_wf = 0
temp = []
for a in alpha:
    df = rockyou_dropnan[rockyou_dropnan.loc[:,'random_guess'] <= a + 0.001]
    sum_random_guess = df.loc[:,'random_guess'].sum()
    g = (1 - df.iloc[-1]['beta_success_rate']) * df["beta_success_rate"].iloc[-1] + sum_random_guess
    result_rockyou.append(g)
```
